```python
import pandas as pd
import os

# --- STEP 1: CREATE 'books.csv' (To ensure the code runs without error) ---
file_name = "books.csv"
if not os.path.exists(file_name):
    data = {
        'Title': ['Python Basics', 'Data Science', 'Machine Learning', 'The Great Gatsby', 'Deep Learning'],
        'Name of Author': ['Guido', 'John Doe', 'Jane Smith', 'F. Scott', 'Jane Smith'],
        'Edition': [1, 3, 2, 10, 4],
        'Publication Year': [2010, 2021, 2018, 1925, 2023],
        'Publishing House': ['OReilly', 'Pearson', 'OReilly', 'Scribner', 'Pearson'],
        'Price': [500, 1200, 1500, 300, 2500]
    }
    pd.DataFrame(data).to_csv(file_name, index=False)
    print("✅ Created sample 'books.csv'")

# --- STEP 2: LOAD AND PERFORM OPERATIONS ---
df = pd.read_csv(file_name)

# a) Print complete report in Tabular form
print("\n--- (a) Complete Book Report ---")
display(df)

# b) Books of a given author
author_name = input("\nEnter Author Name to search (e.g., Jane Smith): ")
print(f"\n--- (b) Books by {author_name} ---")
display(df[df['Name of Author'].str.lower() == author_name.lower()])

# c) Books of a given publishing house
pub_house = input("\nEnter Publishing House to search (e.g., OReilly): ")
print(f"\n--- (c) Books by {pub_house} ---")
display(df[df['Publishing House'].str.lower() == pub_house.lower()])

# d) Titles of cheapest & costliest book
print("\n--- (d) Price Extremes ---")
cheapest = df.loc[df['Price'].idxmin(), 'Title']
costliest = df.loc[df['Price'].idxmax(), 'Title']
print(f"Cheapest Book: {cheapest} (₹{df['Price'].min()})")
print(f"Costliest Book: {costliest} (₹{df['Price'].max()})")

# e) Sorting based on year of publication
print("\n--- (e) Books Sorted by Year (Newest First) ---")
display(df.sort_values(by='Publication Year', ascending=False))
```

```python
import pandas as pd

# --- STEP 1: Create Table of 5 States ---
data = {
    'State': ['Maharashtra', 'Goa', 'Uttar Pradesh', 'Rajasthan', 'Sikkim'],
    'Area': [307713, 3702, 240928, 342239, 7096], # in sq km
    'Population': [112374333, 1458545, 199812341, 68548437, 610577]
}
df_states = pd.DataFrame(data)

# a) Print complete information
print("--- (a) State Information ---")
display(df_states)

# b) State having largest Area
largest_area_state = df_states.loc[df_states['Area'].idxmax(), 'State']
print(f"\n(b) State with largest Area: {largest_area_state}")

# c) State having largest population
largest_pop_state = df_states.loc[df_states['Population'].idxmax(), 'State']
print(f"(c) State with largest Population: {largest_pop_state}")

# d) Calculate population density (Population / Area)
df_states['Density'] = df_states['Population'] / df_states['Area']
print("\n--- (d) Table with Population Density ---")
display(df_states)

# e) State with highest population density
highest_density_state = df_states.loc[df_states['Density'].idxmax(), 'State']
print(f"\n(e) State with highest Population Density: {highest_density_state}")
```
