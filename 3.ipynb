```python
print("--- TASK-1: Patterns ---")

# 1. Row-based increment (1, 12, 123...)
print("\nPattern 1:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()

# 2. Repeated digits (1, 22, 333...)
print("\nPattern 2:")
for i in range(1, 6):
    for j in range(i):
        print(i, end="")
    print()

# 3. Decreasing row numbers (1, 21, 321...)
print("\nPattern 3:")
for i in range(1, 6):
    for j in range(i, 0, -1):
        print(j, end="")
    print()

# 4. Binary/Modulo pattern (1, 10, 101...)
print("\nPattern 4:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j % 2, end="")
    print()

# 5. Continuous even numbers
print("\nPattern 5 (Even Triangle):")
num = 2
for i in range(1, 5):
    for j in range(i):
        print(num, end=" ")
        num += 2
    print()

# 6. Star pattern
print("\nPattern 6:")
for i in range(1, 6):
    print("*" * i)
```

```python
print("--- TASK-2: String Patterns ---")

# 1. Alphabet Triangle (A, AB, ABC...)
print("\nPattern 1:")
for i in range(1, 6):
    for j in range(i):
        print(chr(65 + j), end="") # 65 is 'A'
    print()

# 2. Alternating Star and Hash
print("\nPattern 2:")
for i in range(1, 6):
    for j in range(i):
        if j % 2 == 0:
            print("*", end=" ")
        else:
            print("#", end=" ")
    print()

# 3. Repeated Word Characters (p, yy, ttt...)
print("\nPattern 3:")
word = "python"
for i in range(len(word)):
    print(word[i] * (i + 1))

# 4. "python" Slicing Pattern
print("\nPattern 4:")
for i in range(1, len(word) + 1):
    print(word[:i])
```
